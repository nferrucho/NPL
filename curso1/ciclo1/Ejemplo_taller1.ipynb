```python
from bs4 import BeautifulSoup

def delete_words(soup, id, min_len, max_len, stops):
    # Función simulada para eliminar palabras no deseadas
    text = soup.find(id=id).get_text()
    filtered_text = ' '.join(
        word for word in text.split()
        if min_len <= len(word) <= max_len and word.lower() not in stops
    )
    return filtered_text

def count_words(soup, id, min_len, max_len, stops, words):
    # Utiliza la función delete_words para obtener el texto filtrado
    filtered_text = delete_words(soup, id, min_len, max_len, stops)

    # Inicializa un diccionario para almacenar el conteo de las palabras
    counts = {}

    # Itera sobre la lista de palabras deseadas (words)
    for word in words:
        # Cuenta las ocurrencias de cada palabra en el texto filtrado
        counts[word] = filtered_text.count(word)

    return counts

def print_counts(soup, id, min_len, max_len, stops, words):
    # Obtiene los conteos de palabras
    counts = count_words(soup, id, min_len, max_len, stops, words)

    # Encabezados de la tabla y ancho definido
    header = f"|{'Palabra':<20}|{'Conteo':<20}|"
    separator = "|" + "-" * 20 + "|" + "-" * 20 + "|"

    # Inicializa la tabla con el encabezado y separador
    table = [header, separator]

    # Llena la tabla con los conteos
    for word, count in counts.items():
        table.append(f"|{word:<20}|{count:<20}|")

    # Une la tabla en un solo string para un mejor formato
    result = "\n".join(table)

    return result

# Ejemplo de uso
html_doc = """
<div id="section1">
    Chomshy is a great syntax. Chomshy, syntax, and more syntax.
    Here is more text for testing the syntax of Chomshy. Chomshy is an example word.
</div>
"""

soup = BeautifulSoup(html_doc, 'html.parser')
stops = ['is', 'a', 'and', 'more', 'for', 'the']
words = ['chomshy', 'sintaxis']

# Ejecutar la función
output = print_counts(soup, 'section1', 3, 10, stops, words)
print(output)
```

```python
%%tutor -s -h 500

import re

def count_occurrences(A, B):
    """
    Cuenta cuántas veces las palabras en la lista A aparecen en la lista B,
    considerando que las palabras en B pueden tener caracteres adicionales como comas.

    Parámetros:
    A (list of str): Lista de palabras a contar.
    B (list of str): Lista de palabras en la que buscar.

    Retorna:
    dict: Un diccionario con las palabras de A como claves y sus conteos en B como valores.
    """
    # Crear un diccionario para almacenar los conteos de las palabras en A
    counts = {word: 0 for word in A}

    # Función interna para limpiar palabras
    def clean_word(word):
        return re.sub(r'[^\w\s]', '', word).lower()

    # Limpiar y contar las ocurrencias de cada palabra en la lista B
    for word in B:
        cleaned_word = clean_word(word)
        if cleaned_word in counts:
            counts[cleaned_word] += 1

    return counts

A = ["chomsky", "sintaxis"]
B = ['wikipedia,', 'libre', 'navegación', 'búsqueda', 'noam', 'chomsky', 'noam',
     'chomsky', 'diciembre', '1928', '(93', 'estados', 'william', 'chomsky', 'cónyuge',
     'carol', 'chomsky', 'wasserman', 'chomsky', '(desde', '2014)', 'instituto', 'lane', 'day',
     'high', 'school', '(hasta', '(b.a.;', '(m.a.;', 'harvard', '1951-1955)', 'alumno',
     'zellig', 'goodman', 'filósofo,', 'lingüista,', 'escritor', 'político,', 'profesor', 'psicólogo,', 'activista', 'los', 'derechos',
     'humanos,', 'pedagogo,', 'crítico', 'medios,', 'escritor,', 'teórico', 'área', 'filosofía', 'lenguaje,', 'gramática', 'teoría', 'ciencia', 'cognitiva,',
     'filosofía', 'mente,', 'ética', 'política', 'conocido', 'gramática', 'gramática', 'universal,', 'gramática', 'rección', 'teoría',
     "x',", 'jerarquía', 'chomsky,', 'gramática', 'libre', 'contexto,', 'principios', 'lenguaje,', 'programa', 'pobreza', 'estímulo,',
     'teorema', 'forma', 'normal', 'chomsky,', 'modelo', 'propaganda', 'los', 'medios', 'ocupados', '(desde', '1955)', 'empleador', 'instituto', 'doctorales', ]


result = count_occurrences(A, B)
print(result)
```
